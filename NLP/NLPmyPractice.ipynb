```python
import pandas as pd
```

```python
dataset = pd.read_csv('../NLP/SMSSpamCollection', sep='\t', names=['label', 'message'])
```

```python
dataset.head()
```

```python
dataset['label'].value_counts()/dataset.shape[0]
```

```python
from nltk.corpus import stopwords
```

```python
stopword = stopwords.words('english')
```

```python
import re 
import string
```

```python
string.punctuation
```

```python
def preprocess_text(text):
    # Remove punctuation using regex, convert to lowercase
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text.lower())
    # Tokenize the text
    words = text.split()
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    # Join the words back to form the processed text
    processed_text = ' '.join(words)
    return processed_text
```

```python
dataset['message'] = dataset['message'].apply(preprocess_text)
```

```python
dataset
```

```python
messeges = dataset['message'].values
lables = dataset['label'].values
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
vectorizer = CountVectorizer()
messeges = vectorizer.fit_transform(messeges)
```

```python
messeges.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labencode = LabelEncoder()
lables = labencode.fit_transform(lables)
```

```python
lables
```

```python
messeges_train, messeges_test, lables_train, lables_test = train_test_split(messeges, lables, test_size=0.2, random_state=42)
```

```python
mltnb_clf = MultinomialNB()
```

```python
mltnb_clf.fit(messeges_train, lables_train)
```

```python
y_pred = mltnb_clf.predict(messeges_test)
```

```python
print (accuracy_score(lables_test, y_pred))
print (classification_report(lables_test, y_pred, target_names=labencode.classes_))
```
