# Jaccard Similarity

```python
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    similarity = intersection / union
    return similarity
```

```python
# Example 1:
set_a = set([1, 2, 3, 4, 5])
set_b = set([3, 4, 5, 6, 7])

similarity = jaccard_similarity(set_a, set_b)
print(f"Jaccard Similarity: {similarity}")
```

```python
# Example 2:
set_a = set([1, 1, 1, 1, 1])
set_b = set([1, 1, 1, 1, 1])

similarity = jaccard_similarity(set_a, set_b)
print(f"Jaccard Similarity: {similarity}")
```

# Jaccard Similarity with sklearn

```python
from sklearn.metrics import jaccard_score

# Example usage:
set_a = set([1, 2, 3, 4, 5])
set_b = set([3, 4, 5, 6, 7])

# Convert sets to lists for compatibility with jaccard_score
list_a = list(set_a)
list_b = list(set_b)

# Create binary arrays representing the presence or absence of elements in the sets
binary_array_a = [1 if i in set_a else 0 for i in range(max(max(list_a), max(list_b)) + 1)]
binary_array_b = [1 if i in set_b else 0 for i in range(max(max(list_a), max(list_b)) + 1)]

# Calculate Jaccard similarity using sklearn's jaccard_score
similarity = jaccard_score(binary_array_a, binary_array_b)

print(f"Jaccard Similarity: {similarity}")
```

## Cosine Similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Two lines of text
text1 = "I love this action movie."
text2 = "This is movie is awful"

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Transform the text data into TF-IDF vectors
tfidf_matrix = vectorizer.fit_transform([text1, text2])

# Calculate cosine similarity
cosine_similarity_value = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

# Display the original text
print("Text 1:", text1)
print("Text 2:", text2)

# Display the cosine similarity value
print("\nCosine Similarity:", cosine_similarity_value)
```

```python
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import train_test_split
from surprise.model_selection import cross_validate

# Sample data
data = {
    'user': ['User1', 'User1', 'User2', 'User2', 'User3', 'User3', 'User4', 'User4'],
    'item': ['Item1', 'Item2', 'Item1', 'Item3', 'Item2', 'Item3', 'Item3', 'Item4'],
    'rating': [4, 5, 3, 2, 4, 1, 5, 3],
}

df = pd.DataFrame(data)
df
```

```python
# Define the reader
reader = Reader(rating_scale=(1, 5))

# Load the dataset
dataset = Dataset.load_from_df(df[['user', 'item', 'rating']], reader)

# Split the dataset into a training set and a test set
trainset, testset = train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
# Build the user-based collaborative filtering model using k-NN
sim_options = {
    'name': 'cosine',
    'user_based': True
}

model = KNNBasic(sim_options=sim_options)

# Evaluate the model
cross_validate(model, dataset, measures=['RMSE', 'MAE'], cv=3, verbose=True)

model.fit(trainset)
```

```python
# Make predictions on the test set
predictions = model.test(testset)

# Evaluate the predictions
from surprise import accuracy
accuracy.rmse(predictions)
```

```python
# Make predictions for a given user
user_id = 'User1'
item_id = 'Item3'
rating = model.predict(user_id, item_id).est

print(f"Predicted rating for {user_id} on {item_id}: {rating}")
```
