```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = ["age", "workclass", "fnlwgt", "education", "education_num", "marital_status",
                "occupation", "relationship", "race", "sex", "capital_gain", "capital_loss",
                "hours_per_week", "native_country", "income"]
data = pd.read_csv(url, names=column_names, skipinitialspace=True)
```

```python
# Define the conditions for filtering the data
conditions = [
    (data['relationship'] == 'Unmarried'),
    (data['age'].between(0, 30)),
    (data['marital_status'] == 'Never-married'),
    (data['race'] == 'White'),
    (data['sex'] == 'Female')
]

# Apply the conditions to the data using boolean indexing
filtered_data = data[np.all(conditions, axis=0)]
filtered_data
```

```python
sales_all = walmart['weekly_sales'].sum()
sales_all
```

```python
sales_A = walmart[walmart['type'] == 'A']['weekly_sales'].sum()
sales_A
sales_B = walmart[walmart['type'] == 'B']['weekly_sales'].sum()
sales_C = walmart[walmart['type'] == 'C']['weekly_sales'].sum()
sales_proportions_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_proportions_by_type)
```

```python
sales_by_type = walmart.groupby(['type'])['weekly_sales'].sum()
sales_props = sales_by_type / sum(sales_by_type)
print(sales_props)

```
